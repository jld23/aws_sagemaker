```python
import boto3
import numpy as np
import pandas as pd
import sagemaker
import csv

session = sagemaker.Session()
bucket = 'data-science-on-aws-applied'
role = sagemaker.get_execution_role()
region = boto3.Session().region_name
```

```python
df = pd.read_csv(f's3://{bucket}/customer-reviews/tsv/amazon_reviews_us_Digital_Software_v1_00.tsv.gz',
                delimiter ='\t',
                quoting=csv.QUOTE_NONE,
                compression="gzip")
df.head()
```

```python
df.shape
```

```python
df_grouped = df.groupby('star_rating')
df_balanced = df_grouped.apply(lambda x: x.sample(df_grouped.size().min()).reset_index(drop=True))

df_balanced = df_balanced.reset_index(drop=True)
print(df_balanced.shape)
```

```python
df_balanced.head()
```

```python
from sklearn.model_selection import train_test_split

balanced_train, balanced_holdout = train_test_split(df_balanced, test_size=0.20, stratify=df_balanced['star_rating'])
balanced_validation, balanced_test = train_test_split(balanced_holdout, test_size=0.50, stratify=balanced_holdout['star_rating'])

df_train, df_holdout = train_test_split(df, test_size=0.20, statify=df['star_rating'])
df_validation, df_test = train_test_split(df, test_size=0.50, stratify=df_holdout['star_rating'])
```
